```python
import numpy as np
import pandas as pd 
import geopandas as gpd
```

```python
file_name = r"C:\Users\Kanishk Goyal\OneDrive - IIT Kanpur\Desktop\Prof. Hamim Zafar\Data\Shape file for Districts India\output.shp"
gdf = gpd.read_file(file_name)
```

```python
statename = 'Uttar Pradesh'
my = gdf[gdf.statename == statename]
dist_list = pd.Series(my['distname'])
pd.set_option('display.max_columns', None)
dist_list_values = dist_list.tolist()
```

```python
# Create an empty list to store dictionaries
data_list = []

# Iterate over the list of district names
for dist_name in dist_list_values:
    # Select the geometry for the current district
    district_shape = my[my["distname"] == dist_name].geometry
    
    # Check if the district is found
    if not district_shape.empty:
        # Calculate the centroid
        centroid = district_shape.centroid
        latitude, longitude = centroid.y.values[0], centroid.x.values[0]
        
        # Append the results to the list
        data_list.append({
            "DistrictName": dist_name,
            "Latitude": latitude,
            "Longitude": longitude
        })
    else:
        print(f"District '{dist_name}' not found.")

# Create a DataFrame from the list of dictionaries
result_df = pd.DataFrame(data_list)
```
