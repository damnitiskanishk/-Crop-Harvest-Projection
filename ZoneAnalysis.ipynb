```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

data = df.copy()
```

```python
df['state'].unique()
```

```python
custom_palette = {
    'Karnataka': '#2ca02c',  # green
    'Maharashtra': '#1f77b4',  # blue
    'MadhyaPradesh': '#ff7f0e',  # orange
    'Rajasthan': '#d62728',  # red
    'Uttar Pradesh': '#9467bd'   # purple
}
```

```python
# Select the features to be included in PCA
features = ['rain', 'tmax', 'tmin']
# features = ['rain', 'tmax', 'tmin','year']

# Standardize the features before applying PCA
df_standardized = df.copy()
df_standardized[features] = (df[features] - df[features].mean()) / df[features].std()

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_standardized[features])

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['state'] = df['state']
```

```python
# Initialize the plot
plt.figure(figsize=(10,8))

# Plotting with seaborn
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='state', palette=custom_palette, s=50)

# Adding titles and labels
plt.title('PCA of Rain, Tmax, and Tmin by State')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='State')
plt.grid(True)

# Show plot
plt.show()
```
