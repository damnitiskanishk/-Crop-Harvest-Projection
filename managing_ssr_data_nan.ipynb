```python
df_1 = pd.read_csv(r"C:\Users\Kanishk Goyal\OneDrive - IIT Kanpur\Desktop\Prof. Hamim Zafar\Data\Regridding Data\Rajasthan\Chickpea\ssr_97-22.csv")
df_district = pd.read_csv(r"C:\Users\Kanishk Goyal\OneDrive - IIT Kanpur\Desktop\Prof. Hamim Zafar\Data\Rajasthan_centroid_lat_lon.csv")

df_1 = df_1.drop(columns=['Unnamed: 0'])
df_district.rename(columns={'Latitude':'latitude' , 'Longitude':'longitude', 'Year':'year'}, inplace=True)
```

```python
# Create a list of years
years = []
for year in range(1997, 2022):
    years.append(f"{year}-{year+1}")

# Repeat each row 25 times and add a new 'Year' column
df_2 = pd.DataFrame()
for index, row in df_district.iterrows():
    temp_df = pd.DataFrame({'DistrictName': [row['DistrictName']] * 25,
                            'latitude': [row['latitude']] * 25,
                            'longitude': [row['longitude']] * 25,
                            'year': years})
    df_2 = pd.concat([df_2, temp_df], ignore_index=True)
```

```python
df_2 = df_2.sort_values(by=['DistrictName','year'])
df_2 = df_2.reset_index(drop=True)
```

```python
df_2
```

```python
# Round latitude and longitude in df2 to 6 decimal places
df_district['latitude_rounded'] = df_district['latitude'].round(6)
df_district['longitude_rounded'] = df_district['longitude'].round(6)

district_mapping = dict(zip(zip(df_district['latitude_rounded'], df_district['longitude_rounded']), df_district['DistrictName']))

df_1['latitude_rounded'] = df_1['latitude'].round(6)
df_1['longitude_rounded'] = df_1['longitude'].round(6)

# Map district_name based on rounded latitude and longitude in df1
df_1['DistrictName'] = df_1.apply(lambda row: district_mapping.get((row['latitude_rounded'], row['longitude_rounded']), None), axis=1)

df_1 = df_1.drop(columns=['latitude_rounded','longitude_rounded'])
```

```python
df_1 = df_1.sort_values(by=['DistrictName','year'])
df_1 = df_1.reset_index(drop=True)
```
